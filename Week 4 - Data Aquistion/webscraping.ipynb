```python
import requests
from bs4 import BeautifulSoup
```

```python
# specify the url
url = "https://webscraper.io/test-sites/e-commerce/allinone"
# query the website and return the html to the variable ‘page’
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup)
```

```python
print(soup.find("p", {"class" : "description card-text"}).text)
```

```python
product = soup.find_all("p") #.get_text(strip=True)
print(product)
for i in product:
    print(i.text)
```

```python
title = soup.find("a", {"class" : "title"}).text
price = float(soup.find("h4", {"price float-end card-title pull-right"}).text.replace("$", ""))
desc = soup.find("p", {"class" : "description card-text"}).text

dict = {
    "title": title,
    "price": price,
    "description": desc
    
}
# print(title, price, "\n", desc)
print(dict["price"])
# 
```

```python
import json
with open("laptops.json", "w") as outfile:
    json.dump(dict, outfile, indent=4)
```

```python
## make dict list
dict = {}
lis = []
for i in range(3):
    dict = {
        'title': soup.findAll("a", {"class" : "title"})[i].text.replace("...", ""),
        'price': (float(soup.findAll("h4", {"price float-end card-title pull-right"})[i].text.replace("$", ""))),
        'description': (soup.findAll("p", {"class" : "description card-text"})[i].text)
    }
    lis.append(dict)
    
    


```

```python
import json
with open("laptops.json", "w") as outfile:
    json.dump(lis, outfile, indent=4)
```

```python
import csv
with open("laptop.csv", "w") as outfile:
    header_names = lis[0].keys()
    writer = csv.DictWriter(outfile, header_names)
    writer.writeheader()
    writer.writerows(lis)
```
