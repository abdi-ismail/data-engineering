# Two-Sample T-Test

You’re working in a machine learning team to evaluate the performance of a
recommendation system. You have 2 versions of the model: the original version
(Model A) deployed on dataset 1, and an updated version (Model B) on dataset 2.

Each model was tested on different datasets with similar characteristics and accuracy was recorded. Find out if there’s a statistically significant difference in performance (or accuracy) between Model A and Model B using Python.

1.   Are they related or independent? (Which T-Test are you using?)
2.   What hypothesis are you testing?
3.   Visualise the accuracy distributions for each model





```python
import numpy as np
from scipy.stats import * # Your choice of T Test here!
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a_accuracy = [0.82, 0.85, 0.84, 0.81, 0.83, 0.82, 0.84, 0.85]
b_accuracy = [0.78, 0.79, 0.77, 0.76, 0.80, 0.79, 0.78, 0.77]
```

Two-Sample T-Test

```python
tstat, pval = ttest_ind(a_accuracy, b_accuracy)

print(f"tstat: {tstat}\n",
      f"pval: {pval}")
```

If p-value ≤ α:

This means the result is statistically significant.
We reject the null hypothesis (H₀).
There is sufficient evidence to support the alternative hypothesis (H₁).

If p-value > α:

This means the result is not statistically significant.
We fail to reject the null hypothesis (H₀).
There isn’t enough evidence to support the alternative hypothesis (H₁).

```python
if pval <= 0.05:
    print("Result is statistically significant")
else:
    print("Result is not statistically significant")
```

```python
t_stat, p_value = ttest_rel(a_accuracy, b_accuracy)
print(f"tstat: {tstat}\n",
      f"pval: {p_value}")
```

```python
sns.kdeplot(a_accuracy)
sns.kdeplot(b_accuracy)
plt.show()
```

ANOVA (Analysis of Variance)


```python
np.random.seed(0)
line_a_defects = np.random.normal(5, 1, 30)
line_b_defects = np.random.normal(5.5, 1, 30) 
line_c_defects = np.random.normal(6, 1, 30)  
```

```python
f_stat, p_value = f_oneway(line_a_defects, line_b_defects, line_c_defects)
print(f"fstat: {f_stat}\n",
      f"pval: {p_value}")
```

```python
sns.boxplot(data=[line_a_defects, line_b_defects, line_c_defects])
plt.xticks([0, 1, 2], ['line_a_defects', 'line_b_defects', 'line_c_defects'])
plt.show()
```
