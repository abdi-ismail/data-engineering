```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_palette("muted")
```

```python
df = pd.read_csv("Machines.csv")
```

```python
beginner = df[df["Operator Experience"] == "Beginner"]
intermediate = df[df["Operator Experience"] == "Intermediate"]
expert = df[df["Operator Experience"] == "Expert"]
```

```python
col = "Cycle Time (s)"
# col = "Defects Count"
# col = "Material Thickness (mm)"
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,5))
sns.scatterplot(beginner[col], color="blue", ax=axes[0, 0])
sns.scatterplot(intermediate[col], color="red", ax=axes[0, 1])
sns.scatterplot(expert[col], color="green", ax=axes[1, 0])
```

```python
sns.kdeplot(data=beginner[col], label="beginner")
sns.kdeplot(data=intermediate[col], label="intermediate")
sns.kdeplot(data=expert[col], label="expert")
plt.legend()
```

```python
for i in beginner, intermediate, expert:
    # You fill in the rest!
    shapiro_stat, shapiro_p = stats.shapiro(i[col])

    print(f"\nstat: {shapiro_stat}\n",
      f"p: {shapiro_p}")
    if shapiro_p <= 0.05:
      print("Result is statistically significant\nReject the null hypothesis\n")
    else:
      print("Result is not statistically significant\nAccept the null hypothesis\n")
```

```python
f_stat, p_value = stats.f_oneway(beginner[col], intermediate[col], expert[col])

print(f"\fstat: {f_stat}\n",
      f"p: {p_value}")
```
