```python
form_wins = df[df['Result'] == 'W'].groupby('Formation').size().reset_index(name='Wins').sort_values(by="Wins", ascending=False)
form_wins
```

```python
form_used = df.groupby('Formation').size().reset_index(name='Times Used').sort_values(by="Times Used", ascending=False)

formandwins = pd.merge(form_wins, form_used, on='Formation')

formandwins["Win Rate"] = formandwins["Wins"] / formandwins["Times Used"]

formandwins = formandwins.sort_values(by="Win Rate", ascending=False)
formandwins
```

```python
matchweek_goals = df.groupby("Round")["GF"].sum().reset_index(name="Goals")

```
