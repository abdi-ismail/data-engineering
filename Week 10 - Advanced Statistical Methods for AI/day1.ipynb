```python
import random
from collections import Counter

list = []
turns = 1000000
for i in range(1,turns+1):
    x = random.randint(1,6)
    y = random.randint(1,6)
    total = x + y
    list.append(total)    
```

```python
sns.barplot(Counter(list))
plt.title("Distribution of Rolling Two Dice in 1 Million Simulation")
plt.xlabel("Sum of Two Dice")
plt.ylabel("Count")
plt.show()
```
