```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# List of possible values
poss_values = [1, 2, 3, 4, 5]

# Equal probabilities for each value (uniform distribution)
probabilities = [1/len(poss_values)] * len(poss_values)

# Create the discrete uniform distribution
discrete_uniform = stats.rv_discrete(name='uniform', values=(poss_values, probabilities))

# Sample 100 values
selections = 1000000
samples = discrete_uniform.rvs(size=selections)

# Plot the barplot of the samples
sns.barplot(x=poss_values, y=[np.sum(samples == val) for val in poss_values])
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Discrete Uniform Distribution (100 Samples)')
plt.show()
```
